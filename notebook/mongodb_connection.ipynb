```python
import pandas as pd
from pymongo import MongoClient
```

```python
# Read the heart failure dataset from CSV file
df = pd.read_csv("heart_failure_clinical_records_dataset.csv")
```

```python
# Display first 5 rows to understand the data
df.head(5)
```

```python
# Check the shape of the dataset (rows, columns)
df.shape
```

###  Convert DataFrame to dictionary format

```python
# MongoDB accepts data in dictionary (JSON-like) format
# orient="records" converts each row into a dictionary

data  = df.to_dict(orient="records")
data
```

```python
# Check total number of records
len(data)
```

```python
#MongoDB connection details
DB_NAME ="heart_failure"
COLLECTION_NAME = "heart_data"

#MongoDB Atlas connection URL
CONNECTION_URL = "Use Your Own Url"
```

```python
# Connect to MongoDB

# Create MongoDB client
client = MongoClient(CONNECTION_URL)

# Access the database
data_base = client[DB_NAME]

# Access the collection
collection  = data_base[COLLECTION_NAME]

# Insert multiple records into the collection
insert_result = collection.insert_many(data)
```

```python
# Fetch all records from the collection
records = collection.find()
records
```

```python
# Print first few documents with index
for index, document in enumerate(records):
    print(f"{index}: {document}")
    if index == 4:   # limit output for readability
        break
```

```python
# MongoDB returns cursor  convert it to list  DataFrame
df_mongo = pd.DataFrame(list(collection.find()))
df_mongo.head()
```

```python
# MongoDB automatically adds '_id', which is not needed for ML
if "_id" in df_mongo.columns:
    df_mongo.drop(columns=["_id"], inplace=True)

# Final cleaned DataFrame
df_mongo.head()
```
